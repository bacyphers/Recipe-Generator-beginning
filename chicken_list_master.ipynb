```python
import pandas as pd
import ast 
import isodate
chicken_list=pd.read_csv('/Users/brockberenato/Desktop/capstone - individual overview/chicken_ing')
chicken_list
```

```python
chicken_list.CookTime
```

```python
#all cooktime
def strfdelta(tdelta, fmt):
    d = {"days": tdelta.days}
    d["hours"], rem = divmod(tdelta.seconds, 3600)
    d["minutes"], d["seconds"] = divmod(rem,60)
    day_in_hours= d["days"]*24
    d["hours"] = d["hours"] + day_in_hours
    return fmt.format(**d)
cook_time_list = chicken_list.loc[:, "CookTime"]
for i, cook_time in enumerate(chicken_list['CookTime']):
    new_cook_time = strfdelta(isodate.parse_duration(cook_time), '{hours}:{minutes}')
    chicken_list.at[i, 'CookTime'] = new_cook_time
chicken_list
```
