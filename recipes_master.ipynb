```python
import isodate
def strfdelta(tdelta, fmt):
    d = {"days": tdelta.days}
    d["hours"], rem = divmod(tdelta.seconds, 3600)
    d["minutes"], d["seconds"] = divmod(rem,60)
    day_in_hours= d["days"]*24
    d["hours"] = d["hours"] + day_in_hours
    return fmt.format(**d)
cook_time_list = new_table.loc[:, "CookTime"]

for index, value in cook_time_list.items():
    if isinstance(value, str):
        try:
            isodate.parse_duration(value)
        except isodate.ISO8601Error:
            print(f"Unknown data type encountered at index {index}: {value}")
            break
    else:
        print(f"Unknown data type encountered at index {index}: {value}")
        break
print(cook_time_list[:20])
```
